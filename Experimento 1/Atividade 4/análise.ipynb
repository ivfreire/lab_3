```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Funções

```python
V = lambda t, f, phi, Vp: Vp*np.sin(2*np.pi*f*t + phi)

Ac = lambda A, B, alpha, beta: np.sqrt((A*np.cos(alpha) + B*np.cos(beta))**2 + (A*np.sin(alpha) + B*np.sin(beta))**2)

Phic = lambda A, B, alpha, beta: np.arctan((A*np.sin(alpha) + B*np.sin(beta))/(A*np.cos(alpha) + B*np.cos(beta)))

Vdif = lambda t, f, Vp1, Vp2, phi: Ac(Vp1, -Vp2, phi, 0)*np.sin(2*np.pi*f*t + Phic(Vp1, -Vp2, phi, 0))
```

### Cálculos

```python
t = np.linspace(-0.01, 0.01, 1000)

f = 60

phi = 121*np.pi/180

Vp1 = 16.8

Vp2 = 16.2
```

```python
plt.plot(t, V(t, f, 0, Vp1), label = r'$V_1$')
plt.plot(t, V(t, f, phi, Vp2), label = r'$V_2$')
plt.plot(t, Vdif(t, f, Vp1, Vp2, phi), label = r'$V_{comb}$')
#plt.plot(t, V(t, f, 0, Vp1) - V(t, f, phi, Vp2), label = r'$V_1-V_2$')
plt.legend()
plt.title('Gráfico teórico das tensões por tempo')
plt.ylabel('Tensão (V)')
plt.xlabel('Tempo (s)')
```

```python
plt.plot(t, V(t, f, 0, Vp1), label = r'$V_1$')
plt.plot(t, V(t, f, phi, Vp2), label = r'$V_2$')
#plt.plot(t, Vdif(t, f, Vp1, Vp2, phi), label = r'$V_{comb}$')
plt.plot(t, V(t, f, 0, Vp1) - V(t, f, phi, Vp2), label = r'$V_1-V_2$', c='red')
plt.legend()
plt.title('Gráfico teórico das tensões por tempo')
plt.ylabel('Tensão (V)')
plt.xlabel('Tempo (s)')
```

```python
print(f'Amplitude calculada: {Ac(Vp1, -Vp2, phi, 0)}')
print(f'Fase calculada (graus): {Phic(Vp1, -Vp2, phi, 0)*180/np.pi}')
print(f'Fase calculada (radianos): {Phic(Vp1, -Vp2, phi, 0)}')
```
